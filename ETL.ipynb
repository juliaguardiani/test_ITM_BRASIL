```python
# columns used 
columns02 = ['marca', 'ACA', 'AV', 'BC', 'CH', 'CTE', 'DM', 'discrim', 'ELT', 'EEG', 'EC', 'ES', 'FF', 'HI', 'LDN', 'SR',
             'MEM', 'PHE', 'RR', 'RRP', 'SBF', 'SS', 'subc', 'TC', 'TEC', 'TI', 'usoAgua']

income02 = pd.read_csv(url02, header=None, names=columns02, skiprows=[0])

income02.head()
```

```python
# columns used 
columns03 = ['marca', 'ACA', 'AV', 'BA', 'BC', 'CH', 'CTE', 'DM', 'discrim', 'DI', 'ELT', 'EEG', 'EC', 'ES', 'FF', 
            'HI', 'LDN', 'SR', 'SRR', 'MEM', 'PHE', 'RR', 'RRP', 'SBF', 'SS', 'SMB', 'subc', 'TC', 'TEC', 'TI', 'usuAgua']

income03 = pd.read_csv(url03, header=None, names=columns03, skiprows=[0])

income03.head()
```

```python
for col in df01.columns:
  df01[col] = df01[col].astype(str)
  df01[col] = df01[col].str.replace(',','.')

for col in income02.columns:
  income02[col] = income02[col].astype(str)
  income02[col] = income02[col].str.replace(',','.')

for col in income03.columns:
  income03[col] = income03[col].astype(str)
  income03[col] = income03[col].str.replace(',','.')
```

```python
from sklearn.cluster import KMeans

X_01= df01
kmeans = KMeans(n_clusters=3, random_state=0).fit(X_01)
kmeans.labels_
```

```python
 X_01['label'] = kmeans.predict(X_01)
```

```python
X_01
```

```python
kmeans.cluster_centers_
```

```python
X_01.to_csv('cluster_df01.csv',index=False)
```
